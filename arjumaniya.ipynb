```python
from ultralytics import YOLO
import cv2
import os
import math
```

```python
model = YOLO('yolov8s-pose.pt')
```

```python
input_video_path = 'videos/arjumaniya.mp4'
output_video_path = f'out/pose_{os.path.basename(input_video_path)}'
```

```python
cap = cv2.VideoCapture(input_video_path)
w = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
h = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
out = cv2.VideoWriter(output_video_path, cv2.VideoWriter_fourcc(*'mp4v'), 30, (w, h))
```

```python
count = 0
state = False

while True:
  ret, frame = cap.read()

  if not ret:
    print('Video yakunlandi!!!')
    break

  results = model.predict(frame)

  for result in results:
    for keypoint in result.keypoints:

      x_chap_yelka, y_chap_yelka, conf_chap_yelka = keypoint.data[0][5][0], keypoint.data[0][5][1], keypoint.data[0][5][2]
      cv2.circle(frame, (int(x_chap_yelka), int(y_chap_yelka)), 3, (0, 0, 255), 3)

      x_chap_tirsak, y_chap_tirsak, conf_chap_tirsak = keypoint.data[0][7][0], keypoint.data[0][7][1], keypoint.data[0][7][2]
      cv2.circle(frame, (int(x_chap_tirsak), int(y_chap_tirsak)), 3, (0, 0, 255), 3)

      x_chap_bilak, y_chap_bilak, conf_chap_bilak = keypoint.data[0][9][0], keypoint.data[0][9][1], keypoint.data[0][9][2]
      cv2.circle(frame, (int(x_chap_bilak), int(y_chap_bilak)), 3, (0, 0, 255), 3)

      cv2.line(frame, (int(x_chap_yelka), int(y_chap_yelka)), (int(x_chap_tirsak), int(y_chap_tirsak)), (255, 255, 255), 2)
      cv2.line(frame, (int(x_chap_tirsak), int(y_chap_tirsak)), (int(x_chap_bilak), int(y_chap_bilak)),  (255, 255, 255), 2)

      gradus = math.degrees(math.atan2(y_chap_bilak-y_chap_tirsak, x_chap_bilak-x_chap_tirsak) - math.atan2(y_chap_yelka-y_chap_tirsak, x_chap_yelka-x_chap_tirsak))
      grad = int(360-gradus)

      if state:
        if 150 <= grad:
          count += 1
          state = False

      if grad <= 80:
        state = True

      cv2.putText(frame, f'{str(grad)}', (int(x_chap_tirsak-50), int(y_chap_tirsak)), 1, cv2.FONT_HERSHEY_PLAIN, (0, 0, 0), 2)
      cv2.putText(frame, f'{count}', (100, 100), 2, cv2.FONT_HERSHEY_TRIPLEX, (0, 255, 0), 3)
  out.write(frame)

cap.release()
out.release()
cv2.destroyAllWindows()
```
